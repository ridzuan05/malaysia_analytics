```python
# If table_MN is a list of lists, you can extract the second column values from each list using a list comprehension
column2_values = [row[1] for row in table_MN]
```

```python
column2_values
```

```python
len(column2_values)
```

```python
# flatten a list
flatlist=[element for sublist in column2_values for element in sublist]
print(flatlist)
```

```python
# convert list to dataframe
daerah = pd.DataFrame(flatlist)
```

```python
daerah.head()
```

```python
# rename column with Daerah
daerah = daerah.rename(columns = {0:'Daerah'})
```
